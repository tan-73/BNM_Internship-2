```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding='ISO-8859-1', low_memory = False)
```

```python
df.shape # return rows and columns
```

```python
df.head(10)
```

```python
plt.rcParams['figure.figsize'] = 14,10
sns.countplot(y=df['iyear'], palette = 'inferno')
plt.show()
```

```python
# number of attacks per year

attacks_per_year = df['iyear'].value_counts().sort_index()
plt.figure(figsize=(12,6))
sns.lineplot(x=attacks_per_year.index, y=attacks_per_year.values)
plt.title("Number of Terrorist Attacks Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.tight_layout()
plt.show()
```

```python
# top 10 countries with most attacks

top_countries = df['country_txt'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries with Most Terrorist Attacks")
plt.xlabel("Number of Attacks")
plt.tight_layout()
plt.show()
```

```python
# most common attack types

attack_types = df['attacktype1_txt'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=attack_types.values, y=attack_types.index)
plt.title("Most Common Attack Types")
plt.xlabel("Number of Attacks")
plt.tight_layout()
plt.show()
```

```python
# top 10 deadliest terrorist groups

group_deaths = df.groupby('gname')['nkill'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=group_deaths.values, y=group_deaths.index)
plt.title("Top 10 Deadliest Terrorist Groups")
plt.xlabel("Number of People Killed")
plt.tight_layout()
plt.show()
```
